```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
beml_df = pd.read_csv("BEML.csv")
```

```python
glaxo_df = pd.read_csv("GLAXO.csv")
```

```python
beml_df.head()
```

```python
glaxo_df.head()
```

```python
beml_df = beml_df[['Date','Close']]
```

```python
beml_df
```

```python
glaxo_df = glaxo_df[['Date','Close']]
glaxo_df
```

```python
beml_df.dtypes
```

```python
glaxo_df.dtypes
```

```python
beml_df = beml_df.set_index(pd.DatetimeIndex(beml_df['Date'])) # This will change data coloumninto index
```

```python
glaxo_df = glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Date']))
```

```python
beml_df.head() # now date is acting as an index
```

```python
glaxo_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
plt.plot(glaxo_df.Close)
plt.xlabel('Time')
plt.ylabel('Close Price')
```

```python
plt.plot(beml_df.Close)
plt.xlabel('Time')
plt.ylabel('Close Price')
```

```python
glaxo_df['Gain'] = glaxo_df.Close.pct_change(periods=1) # PCT_CHANGE MEANS PERCENTAGE CHANGE BTEWEEN PRESENTDAY AND PREVIEUS DAY
```

```python
beml_df['Gain'] = beml_df.Close.pct_change(periods=1)
```

```python
glaxo_df.head()
```

```python
beml_df.head()
```

```python
glaxo_df = glaxo_df.dropna()
beml_df = beml_df.dropna()
```

```python
glaxo_df
```

```python
beml_df
```

```python
plt.figure(figsize= (8,5))
plt.plot(glaxo_df.Gain)
plt.xlabel('Time')
plt.ylabel('Gain')
```

```python
plt.figure(figsize= (8,5))
plt.plot(beml_df.Gain)
plt.xlabel('Time')
plt.ylabel('Gain')
```

```python
plt.figure(figsize= (10,5))
plt.suptitle('Distribution', fontsize=(15))
plt.subplot(1,2,1)
sn.distplot(glaxo_df.Gain , label = 'Glaxo' )
plt.xlabel('Gain')
plt.ylabel('Density')
plt.legend()
plt.subplot(1,2,2)
sn.distplot(beml_df.Gain , label = 'BEML' )
plt.xlabel('Gain')
plt.ylabel('Density')
plt.legend()
```

```python
plt.figure(figsize= (10,5))
plt.suptitle('Distribution', fontsize=(15))
sn.distplot(glaxo_df.Gain , label = 'Glaxo' )
sn.distplot(beml_df.Gain , label = 'BEML' )
plt.xlabel('Gain')
plt.ylabel('Density')
plt.legend()
```

```python
print('Mean : ', round(glaxo_df.Gain.mean(), 4))
print('Standard Deviation : ', round(glaxo_df.Gain.std(), 4))
```

```python
print('Mean : ', round(beml_df.Gain.mean(), 4))
print('Standard Deviation : ', round(beml_df.Gain.std(), 4))
```
