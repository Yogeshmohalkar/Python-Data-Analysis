```python
import pandas as pd
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
iris.target_names
```

```python
group1_sentosa = df[df['target']==0 ] ['sepal width (cm)']
group2_versicolor = df[df['target']==1 ] ['sepal width (cm)']
group3_verginica= df[df['target']==2 ] ['sepal width (cm)']
```

```python
group1_sentosa.mean(),group2_versicolor.mean(),group3_verginica.mean()
```

```python
group1_sentosa.hist()
```

```python
group2_versicolor.hist()
```

```python
group3_verginica.hist()
```

```python
import seaborn as sns

sns.displot( data = df,
             x = 'sepal width (cm)',
             hue = 'target',
             fill = True,
             kind = 'kde',
             palette= ['Blue','Orange','Green']
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.scatter(range(1,51) , group1_sentosa)
plt.scatter(range(1,51) , group2_versicolor)
plt.scatter(range(1,51) , group3_verginica)

plt.legend(iris.target_names)
plt.show()
```

```python
from scipy import stats
stats.f_oneway(group1_sentosa,group2_versicolor,group3_verginica)
```

```python
df1 = pd.DataFrame(iris.data)
stats.f_oneway(df1.iloc[:,0],df1.iloc[:,1],df1.iloc[:,2],df1.iloc[:,3])
```
